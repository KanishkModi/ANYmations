- Since the points somewhat make an Elliptical Orbit, the mean coordinates could be taken as the center of the Ellipse.<br>
- The mean of the ordinates of the Highest and Lowest points has been taken as the mean ordinate/Ecliptic Latitude.<br>
- Initially, I thought I would take the 2 points just below the highest point and take their mean to get mean abcissa/Ecliptic Longitude since we dont have the left-most point in the data set.<br>
- However, I observed that the highest and the lowest points of the Ellipse do not give the same abcissa (Clearly due to random error). So, I took the mean of abcissae of the highest and lowest points to get the mean abcissa/Ecliptic Longitude.<br>
- I agree to the fact that this doesn't give the right values of β (mean ecliptic latitude) and λ (mean ecliptic
longitude) but it is a very close approximate and shall not give considerable error while calculating d (mean distance of the star) from them. Also, the least square fitting function of numpy.polyfit doesn't give proper coefficients and the data set already has random error, so the center of Ellipse calculated from the equation of ellipse would also give some error. Therefore I used a much more trivial method to calculate the required data.<br>
- Lastly, I calculated the mean distance "d" of the star as the distance between centre of ellipse and Earth, i.e. by Pythagoras theorem on Mean Ecliptic Latitude and Longitude.<br><p>

```python
import matplotlib.pyplot as plt
import numpy as np

meas = np.genfromtxt('data.csv', delimiter = ',', skip_header = 1)  # Representing given measurements as "meas" by importing using numpy ndarray.

# Creating arrays for Ecliptic latitude and Longitude.
β = meas[:,1]
λ = meas[:,2]

plt.scatter(λ,β,color='k')  # Marks the Data points given on a graph with Black Dots.
plt.plot(λ,β)  # Joins the Data points on the graph with a Blue colored Line.

# Labelling the Graph for better explanation.
plt.xlabel('Ecliptic Longitude λ(t)')
plt.ylabel('Ecliptic Latitude β(t)')
plt.title('Plot of β(t) vs λ(t)')
plt.show()

β_max = max(β)  # To get the ordinate of highest point on the graph.
β_min = min(β)  # To get the ordinate of lowest point on the graph.

# Running a loop to get the coordinates of the highest and lowest point on the graph.
for i in range(len(β)):
    if β[i] == β_max:
        p1 = meas[i]
    elif β[i] == β_min:
        p2 = meas[i]

# Defining the highest and lowest points.
p_high = (p1[1],p1[2])
p_low = (p2[1],p2[2])

# Determining the mean coordinates trivially.
λ_mean = (p2[1] + p1[1])/2
β_mean = (p2[2] + p1[2])/2
print("λ (mean ecliptic longitude) =", λ_mean)
print("β (mean ecliptic latitude) =", β_mean)

d = np.sqrt(λ_mean**2 + β_mean**2)  # To get the mean distance of the star from Earth.
print("d (distance of the star) =", d)
```
