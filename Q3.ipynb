 - I have assumed the larger circle as the light emitting one and the smaller circle to pass in front of it at uniform velocity creating a shadow. <br>
 - Initially, I have taken a distance of 5 units between the circles without overlap and passed the smaller circle of radius "r2 = 6" units along x-axis over the larger circle of radius "r1 = 9" units. <br>
 - Also, I have taken the light intensity emitted by each point as 0.0001. <br>
 - The total time for travelling of the smaller circle is taken to be 420 units. <br>
 - The Plot shown is of the Logarithm of Intensity with time as the smaller circle "c2" passes over the larger circle "c1".

```python
import numpy as np
import matplotlib.pyplot as plt

r2 = 6  # Radius of smaller circle.
r1 = 9  # Radius of larger circle.
n = 2023  # To make a 2023x2023 grid.
a = np.linspace(-30, 30, n)
b = np.linspace(-30, 30, n)
x, y = np.meshgrid(a, b)  # Creating a grid in x and y.

t = np.linspace(0,420,421)  # Taking 421 values of time on x-axis.
Intensity = np.zeros(421)  # Creating an array to be updated later in the loop.

c1 = (x**2 + y**2 <= r1**2).astype(int)  # Taking the larger circle on the grid.

for i in range(0,421,1):
    r = (40/420)*i
    c2 = ((x-r1-r2-5+r)**2 + y**2 <= r2**2)  # Taking the moving smaller circle over the larger circle to create a shadow.
    I = np.logical_and(c1, np.logical_not(c2))  # Creating an array of points emitting light which are not shadowed.
    Intensity[i] = 0.0001*np.count_nonzero(I)  # 0.0001 is taken as the intensity of each point.

plt.plot(t, np.log(Intensity))  # Plotting time on x-axis and Log of Intensity on y-axis.
plt.xlabel('Time')
plt.ylabel('Intensity (Logarithm of Intensity)')
plt.title('Intensity of Emission with Time')
plt.show()
```
