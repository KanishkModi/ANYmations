 -  VARIABLES: v = Recessional velocity of galaxy, d = distance of galaxy, vr = random velocity. <br>
 -  I have taken the component of vr radially outward after considering that the direction of random velocity can be anywhere between 0 to a right angle with the radially outward direction. Since vr is taken as both +ve and -ve, we only take positive projections so as not to have repititve situations. <br>
 -  Assuming Standard Deviation to be 60 km/s which is approximated from that of our own cluster of galaxies. <br>
 -  We take perpendicular velocities and then take their magnitude to get the resultant velocity magnitude of the galaxy. <br>
 -  Using the fact that number density is constant, I took appropriate variation in d as shown by taking cube root of equal spacing till 100^3. Since we move radially outward from a point in 3-D, the number of galaxy would be directly proportional to cube of the distance from the observer.

```python
# Casually importing things required later
import matplotlib.pyplot as plt
import numpy as np
import random as rdm

Ho = 70 # Hubble's constant

d1 = np.linspace(1, 100**3, 420) # To get equal spacing and 420 is a large enough number :P
d = d1**(1/3)

vr = np.array([rdm.gauss(0.0, 60.0) for _ in range(len(d))])
c = np.array([rdm.uniform(0.00, np.pi*0.5) for _ in range(len(d))]) # c = Angle between vr and radially outward direction

v_rec = Ho*d + vr*np.cos(c) # v_rec gives the radially outward velocity of the galaxy
v_tan = vr*np.sin(c) # v_tan gives the velocity of the galaxy normal to the radially outward direction

v = (v_rec**2 + v_tan**2)**0.5  # Gives the magnitude of the net velocity of the galaxy

plt.plot(d, v) 
plt.ylabel('Velocity (magnitude in km/s)')
plt.xlabel('Distance (in Mpc)')
plt.title('Recessional Velocity of Galaxies with Distance')

plt.show()
```
